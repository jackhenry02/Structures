## Part II: Theory

###Equilibrium

If a body is in static equilibrium, then the *vector sum* of all the external forces that act upon it must be zero. This means that there will be no resultant force acting on the body in any direction we might choose to examine. In addition, the *moments* of the applied forces must sum to zero about any point or any line. This is all just Newton’s Second Law, $F = ma$: we want no acceleration, so the (vector) sum of the forces $F$ must be zero. To apply this in practice, we need to decide which part of the universe, which mass $m$, we are talking about. We do not need to know the value of $m$ (as we will only multiply it by $a = 0$) but we must choose a particular mass, draw a boundary around it, and then draw the forces acting on that specific selection. A diagram showing the applied forces acting on the selected component in this manner is called a *free body diagram*.

If we select the lever arm and the load cell as the part of the universe to apply these ideas to, we obtain the free body diagram shown below.  If the lever arm PQ is three times the lever arm QR, apply equilibrium to find the forces P and Q in terms of the force R.

### Pin-jointed frameworks

Triangulated frameworks are widely used in structural engineering for bridges, roof supports, derricks, off-shore oil platforms, space structures etc.; their analysis, and hence their design, can be greatly simplified by idealising the connections between individual members as frictionless hinges, or pin-joints. In a pin-jointed framework with straight members, which carries applied loads only at the joints, the members are either in pure tension or compression. The ideas of equilibrium apply equally well to the framework as a whole or to any part of it.

## Part IV: Derivation of member forces


### Convenience factor

The readings of the strain gauge meters have been calibrated using the expected reading for Bar 1.

We know that the load on the pan is $504\mathrm{lbs} = 229\mathrm{kg} = 2240\mathrm{N}$. Going back to Eqn 1, you hopefully found that $P = R/3$, and so $R = 3P = 3 \times 2240 = 6720 \mathrm{N}$, which is the maximum reading we expect for Bar 1, when the pan is fully off the floor.

The gauge meters in the lab have a 4-digit capacity. However, a minus sign ($-$) takes up a "digit" on the meters. Therefore, we have introduced a "convenience factor" whereby the gauge meters have been calibrated to be $10\times$ less than the reading in $\mathrm{N}$. Therefore, the maximum calibrated reading for Bar 1 is $672$. You simply need to multiply this, and all other readings by $10$, to achieve the value in $\mathrm{N}$.

Note that we are not suggesting that $g=10\mathrm{m/s^2}$, this is simply a "convenience factor" which arose due to the limitations of the gauge meters we have in the lab.

So then why do the readings for Bar 1 go up to just over $700$?

The answer is: the weight of the pan! This was measured for each truss at approximately $11\mathrm{kg}$. Can you think of how we might have measured this in the lab? ### Plot graphs

The `matplotlib.pyplot` object is used to plot graphs of results of two bars against each other. 

As shown in the syntax `df1[reference_bar]`, the square bracket notation conveniently gives the data in a column of a `DataFrame`. 

List comprehension is then used to plot readings from multiple bars against readings from Bar 1.

```python
def plot_raw(reference_bar = 'Bar 1', plot_bars = ['Bar 2', 'Bar 3']):
    plt.figure(figsize=(4.8,4.8))
    [plt.plot(df1[reference_bar], df1[plot_onebar], label = plot_onebar) 
    for plot_onebar in plot_bars]
    plt.xlabel(reference_bar + ' (N)')
    plt.ylabel('Other bars (N)')
    plt.axis('equal')
    plt.legend()
    plt.show()
    
plot_raw()
```

The graph above can be made interactive using `ipywidgets.interact`. The types of interactive elements are specified explicitly here because we want to rename the label (`description`), and `SelectMultiple` does not have a shorthand. The column headings of `df1` are used to automatically (using the convenient feature `.columns`) provide the options of what to plot.

You can use Ctrl+Click or Shift+Click to select multiple bars from the "Bars to plot" selection menu below.

```python
col_names = list(df1.columns)

interact(plot_raw, reference_bar = Dropdown(options = col_names, 
                                            description = 'x-axis'),
          plot_bars = SelectMultiple(options = col_names, 
                                     value = col_names[1:3], 
                                     description = 'Bars to plot'));
```

In this experiment, what matters is not the absolute values, but the changes in readings from the initial values. Therefore, the next section generates the plots after subtracting the initial values.

As shown by the syntax `df1[reference_bar][0]`, the first square bracket retrieves an object containing a column of data from the `DataFrame` using the column heading name; the second square bracket retrieves a single value from the column using an integer position index.

```python
def plot_zeroed(reference_bar = 'Bar 1', plot_bars = ['Bar 2', 'Bar 3']):
    plt.figure(figsize=(4.8,4.8))
    [plt.plot(df1[reference_bar] - df1[reference_bar][0], 
              df1[plot_onebar] - df1[plot_onebar][0], 
             label = plot_onebar
             ) for plot_onebar in plot_bars]
    plt.xlabel(reference_bar + ' (N)')
    plt.ylabel('Other bars (N)')
    plt.axis('equal')
    plt.legend()
    plt.show()

interact(plot_zeroed, reference_bar = Dropdown(options = col_names, 
                                               description = 'x-axis'),
          plot_bars = SelectMultiple(options = col_names, 
                                     value = col_names[1:3], 
                                     description = 'Bars to plot'));
```

### Linear regression
This part of the lab exercise is to verify the theory of static equilibrium with the experimental data. This is done by comparing the changes in results from different bars on the structure. 

The ratio of reading changes is extracted using linear regression. This functionality is provided by `numpy.polyfit`. As shown in the code below, the first input parameter to `polyfit` is for values of $x$, the second input parameter is for values of $y$, and the third parameter is an integer for the order of equation. A linear equation has order 1 because the highest power of $x$ in the equation $y = m x + c$ is 1.

```python
reference_bar = 'Bar 1'
plot_onebar = 'Bar 3'

model = np.polyfit(df1[reference_bar], df1[plot_onebar], 1)
model
```

`polyfit` returns a `numpy.array` with the head of the array being the highest-order coefficient. In other words, `model[0]` gives the slope $m$ of the best-fit straight line.

This can be built into another adaptation of the graph-plotting script from above. The experimental data points are plotted as a scatter graph, the best-fit line is plotted as a solid line, and the best-fit equation is given in the legend.

You can choose which bars to compare in this graph using the interactive drop-down boxes.

```python
def plot_linreg(reference_bar, plot_onebar):
    plt.figure(figsize=(4.8,4.8))
    model = np.polyfit(df1[reference_bar], df1[plot_onebar], 1)

    plt.scatter(df1[reference_bar], df1[plot_onebar], label = 'Expt data')

    c0 = model[1] # Intercept
    c1 = model[0] # Slope
    x0 = min(df1[reference_bar])
    x1 = max(df1[reference_bar])
    y0 = c0 + x0 * c1
    y1 = c0 + x1 * c1
    plt.plot([x0, x1], [y0, y1], label = ("y = %.3f x + %.1f" % (c1, c0)))
    plt.xlabel(reference_bar + ' (N)')
    plt.ylabel(plot_onebar + ' (N)')
    plt.axis('equal')
    plt.legend()
    plt.show()

interact(plot_linreg, 
         reference_bar = Dropdown(options = col_names, value = col_names[0], 
                                  description = 'x-axis'),
         plot_onebar = Dropdown(options = col_names, value = col_names[1], 
                                description = 'y-axis')
        );
```

The next code cell performs linear regression to find the ratio of reading changes between all bars and Bar 1. The script uses list comprehension to loop through all the bars using `for .. in`, filtering out Bar 1 itself using `if .. is not`.

The results of linear regression are then presented in a table, using `DataFrame` again to provide the formatting.

```python
reference_bar = 'Bar 1'
models = [np.polyfit(df1[reference_bar], df1[onebar], 1) for onebar in col_names 
          if onebar is not col_names[0]]

dfc = pd.DataFrame([model[0] for model in models], 
                   index=pd.Index(col_names[1:]), 
                   columns=['Ratio vs ' + reference_bar])
dfc
```

Hopefully these ratios relative to Bar 1 will be close to the theoretical values you predicted by the principle of equilibrium in the 'Predicted' tab of your Google Sheet (allowing for some experimental error, and for the fact that the joints are not perfectly pinned).

###Comparison of predicted and measured results

We will now compare the theoretical predicted load and the experimental loads, rather than just the ratios to Bar 1, as we did above. We will do this using a similar graph to those above, except we will compare our experimental forces to the predicted (or expected) forces, rather than to Bar 1.

We will use the 'Predicted' worksheet/tab of the Google Sheet, which we calculated earlier, and get that into a new `DataFrame`, so we can use it here.

Open **your copy** of the Google Sheet now. We import that worksheet in a similar way to before, changing the code slightly so we copy in the relevant worksheet.

```python
worksheet2 = gc.open('TrussTemplate').worksheet('Predicted') #Rename your Google Sheet here

# get_all_values gives a list of rows.
rows2 = worksheet2.get_all_values()

# Convert to a DataFrame and render.
df2 = pd.DataFrame.from_records(rows2)

df2 = df2.drop([0])
df2 = df2.drop([0], axis=1)

df2 = df2.rename(index={1:'Bar 1', 2:'Bar 2', 3:'Bar 3', 
                        4:'Bar 4', 5:'Bar 5', 6:'Bar 6', 7:'Bar 7', 8:'Bar 8', 
                        9:'Bar 9', 10:'Bar 10', 11:'Bar 11', 12:'Bar 12'})
df2 = df2.rename(columns={1:'Expected load /R'})

df2 = df2.astype('double')
df2
```

We don't have the exact load on the pan (W) for each of our experimental data points, as we loaded and unloaded the truss with the spanner and turnbuckle fairly arbitrarily. However, we can still get our predicted forces in each bar as follows:
1. Use the measured values in Bar 1, and the "convenience" factor, to obtain the load that was in Bar 1 (R) at each stage.
2. Use the table above with the theoretical bar force ratios for the other bars to obtain the predicted forces in all bars.

To do this, we will now create our own `DataFrame`, which will be similar to the measured values one we used above, but will be for predicted values in N.

The code below creates an empty table with column captions and row captions, which are the same as the measured values `DataFrame` above. The `numpy` functions `zeros` is used as a shortcut to generate a two-dimensional array of integer zeros.

In the code cells above we labelled rows and columns manually, here we automate this a bit more. You can adjust the values of `bar_count` and `stage_count` to suit your input data, or even change the code for `row_names` and `col_names` to use different column and row headings.

```python
bar_count = 12
stage_count = 6 # Excludes initial stage
blank_boxes = np.zeros((stage_count+1, bar_count), dtype=int)

row_names2 = ['Initial'] + [("Stage %d" % ii) for ii in range(1, stage_count+1)]
col_names2 = [("Bar %d" % ii) for ii in range(1, bar_count+1)]

df3 = pd.DataFrame(blank_boxes, index=pd.Index(row_names2), columns=pd.Index(col_names2))
df3
```

We can now start to populate this table with our predicted forces. The first step is to copy and paste the measured Bar 1 values.

```python
df3[['Bar 1']] = df1[['Bar 1']]
df3
```

We can now go through and multiply the Bar 1 forces by the theoretical bar force ratios for all the other bars, to obtain a full table of values.

We loop over each column, populating it by multiplying the first column by the appropriate predicted load ratio.

```python
for column in df3.columns[:]:
  df3[column] = (df2.iloc[df3.columns.get_loc(column)]['Expected load /R'])*df3[['Bar 1']]

df3
```

Before you move on, there is one point which we did not load arbitrarily - the maximum load. We know the weights add up to 504lbs and the weight of the pan is about 11kg. 

So, let's check the Stage 4 values, by using the addition of these for $P$ and calculating the true maximum theoretical bar readings for all other bars, as we did above, but in a smaller `DataFrame`. If these agree, we can feel more confident about using the Bar 1 measured values above for the arbitrary loading points we took in the photos. 

We found the theoretical Bar 1 force before as $6720\mathrm{N}$. The weight of the pan is: $11\mathrm{kg} = 108\mathrm{N}$. Eqn 1 converts the load in the pan to the load on Bar 1: $108\times 3 = 324\mathrm{N}$. Finally, the total max load is: $6720 + 324 = 7040\mathrm{N}$ (rounded down to allow for the convenience factor). Therefore, our Bar 1 load is $7040\mathrm{N}$ and we achieve the other bar loads as above.

```python
bar_count = 12
stage_count = 1 #Initial and fully loaded (Stage 4), but excluding initial!
blank_boxes = np.zeros((stage_count+1, bar_count), dtype=int)

row_names3 = ['Initial'] + ['Stage 4']
col_names3 = [("Bar %d" % ii) for ii in range(1, bar_count+1)]

df4 = pd.DataFrame(blank_boxes, index=pd.Index(row_names3), columns=pd.Index(col_names3))

df4[['Bar 1']] = np.transpose([[0,7040]])

for column in df4.columns[:]:
  df4[column] = (df2.iloc[df4.columns.get_loc(column)]['Expected load /R'])*df4[['Bar 1']]

df4
```

If you're satisfied that the Stage 4 maximum expected values above match those calculated using the maximum measured Bar 1 loads, you can now plot some graphs using the values we calculated in `df3`.

We now have all the data we need and can plot a graph of predicted against measured forces, using the dataframes `df3` and `df1`, respectively. Before you run the code below, think about what you might expect this graph to look like?

The code below plots such graphs of predicted vs measured forces for each bar, and you can again interact with the plot and look at each bar in turn using the dropdown. Do the graphs show what you expected? 

```python
def plot_compare(bars = ['Bar 1']): 
    plt.figure(figsize=(4.8,4.8))
    [plt.plot((df1[bars] - df1[bars][0]),
              (df3[bars] - df3[bars][0]),
              label = bars)]
    plt.xlabel('Measured (N)')
    plt.ylabel('Predicted (N)')
    plt.axis('equal')
    plt.show()

interact(plot_compare, bars = Dropdown(options = col_names, 
                                              description = 'Bar'));
```
